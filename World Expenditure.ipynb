```python
import pandas as pd

# Reading the CSV file into a DataFrame
df = pd.read_csv('C:/Users/ASUS/Downloads/WorldExpenditures.csv')

# Display the DataFrame
print(df)
```

```python
#Finding the number of blanks in the each columns within the dataset

blanks_df = df[df['Country'].isnull()]
print(blanks_df)
blanks_df_year = df[df['Year'].isnull()]
print(blanks_df_year)
blanks_df_gdp = df[df['GDP'].isnull()]
print(blanks_df_gdp)
```

```python
#Plotiing data of 96 goverment data across 20 different sectors 

from matplotlib import pyplot as plt
plt.figure(figsize=(20, 8))
sectors = df['Sector'].unique()
for i, sector in enumerate(sectors):
    plt.subplot(5, 4, i+1)  # Create subplots
    for country in df['Country'].unique():
        country_df = df[(df['Sector'] == sector) & (df['Country'] == country)]
        plt.plot(country_df['Year'], country_df['Expenditure(million USD)'], label=country)
    plt.title(sector)
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.legend(loc='upper left')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the data
# You may need to pivot the data to create a suitable format for the heatmap
heatmap_data = df.pivot_table(index='Country', columns='Sector', values='Expenditure(million USD)', aggfunc='sum')

plt.figure(figsize=(20, 10))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5)
plt.title('Expenditure (million USD) by Country and Sector')
plt.xlabel('Sector')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filtering the data to include only the years from 2000 to 2022
filtered_df = df[(df['Year'] >= 2000) & (df['Year'] <= 2022)]

plt.figure(figsize=(12, 8))

# Iterate over unique countries
for country in filtered_df['Country'].unique():
    country_data = filtered_df[filtered_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Expenditure(million USD)'], label=country)

plt.title('Expenditure (million USD) by Year and Country')
plt.xlabel('Year')
plt.ylabel('Expenditure (million USD)')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid()
plt.show()
```
